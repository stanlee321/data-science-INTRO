### Mean vs. Median


Vamos a crear "fake data" , _fake data_ que represente la ganancia promedio de las personas al rededor de un año,  lo vamos a centrar al rededor de 27,000  (osea con una media de 27000), va a ser una  Distribucion Normal y una desviacion estandar de of 15,000, en total van a ser  10,000 data points. 

Vamos a calcula la media (average, valor promedio del data set) - este deberia estar sercano a 27,000:


```python
import numpy as np  # Importamos Numpy que es la lib de calculo numerico de Python


incomes = np.random.normal(27000, 15000, 10000) #  la funcion np.random.normal agarra tres argumentos
                                                #   media, standar desviation, datapoints 

# Se calcula la media  en una simple linea de codigo con ayuda
# de Numpy

mean = np.mean(incomes)
print(mean)
```

Recordemos que la Varianza es la medida de cuanto mas o menos el data esta alejado de su media
Simplemnte es el cuadrado de la resta de todos los valores con la media.
Example:
    cual es la variance del dataset (1, 4 , 5, 4, 8)??
    * Se calcula la media: Media = (1 + 4 + 5 + 4 + 8 ) / 5 = 4.4
    * Se encuentra las diferencias: (-3.4, -0.4, 0.6,-0.4,3.3)
    * Se eleva al cuadrado  cada uno: (11.56, 0.16, 0.36, 0.16, 12.96)
    * Encontrar la suma promedio : 


$\sigma^2$ = (11.56+ 0.16+ 0.36+ 0.16 +12.96) / 5 = 5.04

La Desviacion Estandar es

```python
incomes.std()
```

Standar Desviation es una herramienta para identificar Outliers, al elevar al cuadrado la varianza se obtiene la desviacion std, esta std  amplifica los Outliers y es una manera de reconocerlos.

std  = $ \sqrt(\sigma^2)$

# 8 Covarianza y Correlacion

La covarianza mide cómo dos variables varían de sus valores medios o means.
Por ejemplo, supongamos que trabajamos para una empresa de comercio electrónico, y están interesados en encontrar una correlación entre la velocidad de la página (qué tan rápido carga cada página web para un cliente) y cuánto gasta un cliente.

Numpy ofrece métodos de covarianza, pero lo haremos de la "manera difícil" de mostrar lo que ocurre debajo del sombrero. Básicamente tratamos cada variable como un vector de desviaciones de la media, y calculamos el "producto interno o dot product" de ambos vectores. Geométricamente, se puede considerar que este es el ángulo entre los dos vectores en un espacio de alta dimensión, pero solo se puede considerar como una medida de similitud entre las dos variables.

Primero, hagamos que la velocidad de la página y la cantidad de compra sean totalmente aleatorias e independientes entre sí; se producirá una covarianza muy pequeña ya que no existe una correlación real:

```python
import numpy as np
from pylab import * #importamos todo de pylab incluyendo scatter y covariance

def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

Ahora haremos que nuestros montos de compra inventados sean una función real de la velocidad de la página, haciendo una correlación muy real. El valor negativo indica una relación inversa; las páginas que se renderizan en menos tiempo resultan en más dinero gastado:

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

Pero, ¿qué significa este valor? La covarianza es sensible a las unidades utilizadas en las variables, lo que dificulta la interpretación. Una solucion?, la correlación normaliza todo según sus desviaciones estándar, lo que le proporciona un valor más fácil de comprender que oscila entre -1 (para una correlación inversa perfecta) y 1 (para una correlación positiva perfecta):

```python

def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    
    return covariance(x,y) / stddevx / stddevy  #En la vida real, hay que tener cuidado de no dividir entro 0



correlation(pageSpeeds, purchaseAmount)
```


numpy puede hacer todo esto por ti con numpy.corrcoef. Devuelve una matriz de los coeficientes de correlación entre cada combinación de las matrices pasadas ya normalizado entre -1 a 1, listo para su interpretacion:

```python
np.corrcoef(pageSpeeds, purchaseAmount)
```

(No coincide exactamente solo debido a la precisión matemática disponible en una computadora).
Podemos forzar una correlación perfecta mediante la fabricación de una relación totalmente lineal (una vez más, no es exactamente -1 solo por errores de precisión, pero está lo suficientemente cerca como para decirnos que aquí hay una correlación realmente buena):

```python
purchaseAmount = 100 - pageSpeeds * 3

scatter(pageSpeeds, purchaseAmount)

correlation (pageSpeeds, purchaseAmount)
```


Hay que recordad que  la correlación no implica causalidad.
